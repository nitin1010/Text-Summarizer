```python
from transformers import pipeline
```

```python

model_name = "sshleifer/distilbart-cnn-12-6"
summarizer = pipeline("summarization", model=model_name)
```

```python
text = """
The old house creaked like a ship caught in a storm. It stood at the end of Willow Lane, a solitary sentinel against the encroaching city. Inside, dust motes danced in sunbeams that slanted through grimy panes. It was in this house that Eleanor lived, a woman as old as the house itself, her life etched into the worn wood and peeling paint.
Eleanor was a recluse, her world confined to the house's four walls. Time had softened her edges, turning her once vibrant spirit into a muted watercolor. Days were spent in a rocking chair by the fireplace, a worn book her only companion. Nights were filled with the symphony of creaking floorboards and the whisper of wind through the eaves.
One winter, a blizzard howled outside. The house was a fortress against the storm, but inside, Eleanor felt a chill deeper than the frost. A strange occurrence began. Objects moved, doors creaked open of their own accord. Shadows seemed to dance with an eerie life of their own. Fear, a stranger to Eleanor, crept in.
Then she found it, hidden beneath floorboards in the attic: a diary. It belonged to a young woman, a former occupant of the house. The diary painted a picture of a life filled with passion, heartbreak, and a tragic end. As Eleanor delved deeper into the past, the boundaries between the living and the dead began to blur.
She discovered the young woman's spirit was trapped, unable to move on. The house, it seemed, was a prison as well as a sanctuary. Slowly, a connection formed between them, an unspoken understanding. Eleanor, with her quiet strength, became the young woman's anchor, a bridge to the other side.
Together, they worked to unravel the mysteries of the past. Secrets were uncovered, wounds were healed. The house, once a place of shadows, began to fill with light. And as the first blooms of spring touched the old house, Eleanor felt a sense of peace she hadn't known in decades. The storm within her was subsiding, replaced by a gentle calm. The house still creaked, but the sounds were no longer haunting. They were the rhythm of life, of a past reconciled with the present.
As the world outside blossomed, so too did Eleanor. She found a strength she never knew she possessed. The recluse was slowly transforming, her heart opening to the world beyond her walls. And in the quiet moments, when the house was still, she could almost feel the presence of her ghostly companion, a silent guardian in the tapestry of her life.
"""
```

```python
# Specify desired summary length
desired_summary_length = 300  # Adjust as needed

# Calculate max_length and min_length in tokens
# Assuming average English word length is about 5 characters + space = 6 characters
# Thus, for 300 words, the length in characters would be around 300 * 6 = 1800 characters
max_length = desired_summary_length * 6 // 5
min_length = max_length // 3  # Ensure min_length is one-third of max_length



summary = summarizer(text, max_length=max_length, min_length=min_length, do_sample=False)
summary_text = summary[0]['summary_text']
```

```python
char_count = len(summary_text)
word_count = len(summary_text.split())

print("Original Text:\n", text)
print("\nSummary:\n", summary_text)
print("\nSummary Length:\n", f"Characters: {char_count}, Words: {word_count}")
```
