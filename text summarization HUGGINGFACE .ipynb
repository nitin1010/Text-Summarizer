```python


import torch
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM, pipeline
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.document_loaders import PyPDFLoader

# Load model directly
tokenizer = AutoTokenizer.from_pretrained("MBZUAI/LaMini-Flan-T5-248M")
model = AutoModelForSeq2SeqLM.from_pretrained("MBZUAI/LaMini-Flan-T5-248M")

# Model and tokenizer loading

pipe = pipeline("text2text-generation", model=model, tokenizer=tokenizer)

tokenizer = AutoTokenizer.from_pretrained("MBZUAI/LaMini-Flan-T5-248M")
model = AutoModelForSeq2SeqLM.from_pretrained("MBZUAI/LaMini-Flan-T5-248M")


def preprocess_text(text):
    # Add preprocessing steps here, e.g., removing stop words, punctuation, stemming
    return text

def file_preprocessing(file_path):
    loader = PyPDFLoader(file_path)
    pages = loader.load_and_split()
    text_splitter = RecursiveCharacterTextSplitter(chunk_size=200, chunk_overlap=50)
    texts = text_splitter.split_documents(pages)
    final_texts = ""
    for text in texts:
        final_texts += preprocess_text(text.page_content)  # Apply preprocessing
    return final_texts

def llm_pipeline(file_path):
    summarization_pipeline = pipeline(
        'summarization',
        model=model,
        tokenizer=tokenizer,
        max_length=100,
        min_length=50
    )
    input_text = file_preprocessing(file_path)
    result = summarization_pipeline(input_text)
    return result[0]['summary_text']

# Example usage
file_path = '/content/Operations Management.pdf'  # Update with your PDF file path
summary = llm_pipeline(file_path)
print("Summary:\n", summary)
```
