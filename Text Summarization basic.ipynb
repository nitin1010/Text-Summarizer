```python
text = """Once upon a time, in a green forest, lived a little squirrel named Sammy. Sammy loved acorns more than anything. He'd spend his days scampering up and down trees, searching for the perfect acorn. One autumn, Sammy found a giant acorn, bigger than his head! He tried to carry it home but it was too heavy. A wise old owl, perched on a nearby branch, watched Sammy struggle. With a hoot, the owl suggested Sammy make friends with the other forest creatures to help him carry the acorn. Sammy agreed. He asked the strong bear, the clever fox, and the friendly deer for help. Together, they worked as a team, taking turns carrying the giant acorn. They had a picnic under the biggest oak tree, sharing berries and nuts. As the sun began to set, casting long shadows, they gently placed the acorn in a cozy spot. Sammy thanked his new friends and knew that friendship was as important as finding the perfect acorn."""
```

```python
len(text)
```

```python
import spacy
from spacy.lang.en.stop_words import STOP_WORDS
from string import punctuation
```

```python
nlp = spacy.load('en_core_web_sm')
```

```python
doc = nlp(text)
```

```python
tokens = [token.text for token in doc]
print(tokens)
```

```python
punctuation = punctuation + '\n'
punctuation
```

```python
word_frequencies = {}
stopwords = list(STOP_WORDS)

for word in doc:
    if word.text.lower() not in stopwords:
        if word.text.lower() not in punctuation:

            if word.text not in word_frequencies.keys():
                word_frequencies[word.text] = 1
            else:
                word_frequencies[word.text] += 1
```

```python
print(word_frequencies)
```

```python
max_frequency = max(word_frequencies.values())
max_frequency
```

```python
for word in word_frequencies.keys():
    word_frequencies[word] = word_frequencies[word]/max_frequency
```

```python
print(word_frequencies)
```

```python
sent_tokens = [sent for sent in doc.sents]
print(sent_tokens)
```

```python
sent_score = {}
```

```python
for sent in sent_tokens:
    for word in sent:
        if word.text.lower() in word_frequencies.keys():
            if sent not in sent_score.keys():
                sent_score[sent] = word_frequencies[word.text.lower()]
            else:
                sent_score[sent] += word_frequencies[word.text.lower()]
```

```python
print(sent_score)
```

```python
from heapq import nlargest
```

```python
len(sent_score) * 0.3
```

```python
summary = nlargest(n=12, iterable=sent_score, key=sent_score.get)
print(summary)
```

```python
final_summary = [word.text for word in summary]
final_summary
```

```python
final = " ".join(final_summary)
final
```
